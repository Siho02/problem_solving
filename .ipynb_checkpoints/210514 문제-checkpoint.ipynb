```python
#   
# 91 ~ 100 = A, 81~90 = B, 71 ~ 80 = C, 61~70=D,   F

point = int(input("   : "))

def grade_check(point) :
    if point > 90 :
        grade = 'A'
    elif point > 80 :
        grade = 'B'
    elif point > 70 :
        grade = 'C'
    elif point > 60 :
        grade = 'D' 
    else :
        grade = 'F'
    
    print("", grade,"")
    
grade_check(point)
    
```

```python
ls = list(map(int, input().split(' ')))

def eq_noteq(ls) : 
    for idx in range(len(ls)-1) :
        if ls[idx] != ls[idx+1] :
            continue
        else :
            print("  ",idx,"", idx+1, "   !")
        
print(ls, "   = ", len(ls))
eq_noteq(ls)
```
