```python
# 시험 성적 문제
# 91 ~ 100 = A, 81~90 = B, 71 ~ 80 = C, 61~70=D, 그 이하는 F

point = int(input("시험 점수를 입력하세요 : "))

def grade_check(point) :
    if point > 90 :
        grade = 'A'
    elif point > 80 :
        grade = 'B'
    elif point > 70 :
        grade = 'C'
    elif point > 60 :
        grade = 'D' 
    else :
        grade = 'F'
    
    print("성적은", grade,"입니다")
    
grade_check(point)
    
```

```python
# 설정 온도, 입력 온도 문제

set_temp = float(input("설정 온도를 입력하세요 : "))
input_temp = float(input("입력 온도를 입력하세요 : "))

def temp_comprison(set_temp, input_temp) :
    if set_temp == input_temp :
        print("적정 온도입니다")
    else :
        if set_temp > input_temp :
            print("cold")
        else :
            print("hpt")
            
temp_comprison(set_temp, input_temp)

        
```

```python
ls = list(map(int, input().split(' ')))

def eq_noteq(ls) : 
    for idx in range(len(ls)-1) :
        if ls[idx] != ls[idx+1] :
            continue
        else :
            print("주어진 리스트의 ",idx,"번째와", idx+1, "번째 요소는 서로 같다!")
        
print(ls, "주어진 리스트의 길이 = ", len(ls))
eq_noteq(ls)
```
